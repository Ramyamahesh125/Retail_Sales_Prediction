```python
# Function for choosing best algorithm by R2 Score:

def accuracy_score(df,algorithm):
    
    x = df.drop(columns = ["Weekly_Sales","Differ"]).fillna(0)
    y = df["Weekly_Sales"].fillna(0)
    x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.2, random_state = 42)
    
    model=algorithm().fit(x_train,y_train)
    y_pred_train= model.predict(x_train)
    y_pred_test= model.predict(x_test)

    r2_score_train=r2_score(y_pred_train,y_train)
    r2_score_test= r2_score(y_pred_test,y_test)

    accuracy={"algorithm":algorithm,
              "R2_train":r2_score_train,
              "R2_test":r2_score_test}
    
    return accuracy
```

```python
# Function using Random Forest Regressor:

def RandomForest(df, null_columns, lable):

    # dropping null columns 
    df1 = df.drop(columns = null_columns)

    # Separating null columns for lable:
    df1_null = df1[df1[lable].isnull()]
    df1_null.reset_index(drop = True, inplace = True)

    # separating Not Null columns for lable:
    df1_notnull = df1[df1[lable].notnull()]
    df1_notnull.reset_index(drop = True, inplace =True)

    # choosing x and y value from notnull dataframe:
    x = df1_notnull.drop(columns = [lable, "Differ"])
    y = df1_notnull[lable]

    # Train test splitting:
    x_train, x_test, y_train, y_test = train_test_split(x,y , test_size = 0.2, random_state = 42)

    model = RandomForestRegressor().fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # evaluating the model:
    mse = mean_squared_error(y_pred, y_test)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_pred, y_test)
    r2 = r2_score(y_pred, y_test)

    metrics = {"Mean Squared Error": mse,
               "Root Mean Squared Error": rmse,
               "Mean Absolute Error": mae,
               "R2 Score": r2}
    print(metrics)
    # predict the null values for the df1_null df:
    test_data = df1_null.drop(columns = [lable,"Differ"])
    y_pred_null_df = model.predict(test_data)

    df1_null[lable] = pd.DataFrame(y_pred_null_df)

    df_result = pd.concat([df1_null, df1_notnull], axis = 0, ignore_index= True)

    return df_result
```

```python
# ML function for model 2:

def RandomForest(df,null_columns,lable):

    df1 = df.drop(columns = null_columns)

    # separating null columns:
    df1_null = df1[df1[lable].isnull()]
    df1_null.reset_index(drop = True, inplace = True)

    # separating not null columns:
    df1_notnull = df1[df1[lable].notnull()]
    df1_notnull.reset_index(drop = True, inplace = True)

    # choosing x and y value from notnull  dataframe:
    x = df1_notnull.drop(columns = [lable,"Differ"])
    y = df1_notnull[lable]

    #train test splitting:
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)

    model = RandomForestRegressor().fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # evaluating the model:
    mse = mean_squared_error(y_pred, y_test)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_pred, y_test)
    r2 = r2_score(y_pred, y_test)

    metrics = {"Mean Squared Error":mse,
               "Root Mean Squared Error": rmse,
               "Mean Absolute Error": mae,
               "R2 Score": r2}
    print(metrics)

    # predicting the null values for null dataframe
    test_data = df1_null.drop(columns = [lable, "Differ"])
    y_pred_null_df = model.predict(test_data)

    df1_null[lable] = pd.DataFrame(y_pred_null_df)

    df_result = pd.concat([df1_null,df1_notnull], axis = 0, ignore_index=True)

    return df_result
```
